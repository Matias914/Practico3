## Analizar los valores de las columnas. ¿Qué tipo de datos hay en cada una? ¿Qué representan? Incluir información del análisis en la notebook.

- **SHOW_ID:** dato cualitativo nominal. Sirve a modo de ID y no tiene información de interés ligada a él.
- **TYPE:** dato cualitativo nominal. Se refiere al tipo: pelicula o programa, y esto puede ser útil para saber cuál es la tendencia de la gente.
- **TITLE:** dato cualitativo nominal. Representa el título y puede ser de interés para saber qué tan llamativos son.
- **DIRECTOR:** dato cualitativo nominal. Indica el director de la obra y puede servir los gustos de los espectadores.
- **CAST:** dato cualitativo nominal. Muestra el reparto, lo que ayuda a saber las tendencias porque podrían haber actores célebres o de interés.                     
- **COUNTRY:** dato cualitativo nominal. Indica el origen de la película y su motivo puede ser que las personas de un lugar tienen más predisposición a ver películas locales que los extranjeros.                          
- **DATE_ADDED:** dato cualitativo ordinal. Se refiere a la fecha en la que fue añadido el producto a la plataforma. Si es novedoso podría ser de interés y se podría evaluar cómo reacciona la gente.                            
- **RELEASE_YEAR:** dato cuantitativo discreto. La fecha de lanzamiento tiene que ver con qué tan anticuada es una obra para una sociedad actual (por dar un ejemplo)                          
- **RATING:** dato cualitativo ordinal. El rating del filme es importante para saber los gustos de la gente y de la crítica.                        
- **DURATION:** dato cuantitativo continuo. Brinda información sobre la capacidad de atención de la gente y su predispoción a ver películas largas o cortas.                           
- **LISTED_IN:** dato cualitativo nominal. Habla de como fue clasificada la obra dentro de la obra. Lo mismo que fue aclarado, brinda información sobre tendencias.                            
- **DESCRIPTION:** dato cualitativo nominal. Se podría evaluar el impacto de la sinopsis, por ejemplo, qué tan llamativa.

## Responder las siguientes preguntas:

#### I) ¿Cuántos directores distintos hay?

```python
frecuencia_directores = raw_dataframe["director"].value_counts() 

num_directores = frecuencia_directores.shape[0]
print(f"La cantidad de directores: {num_directores}\n")
print(f"Los directores en orden según la cantidad de veces que aparecen:\n\n{frecuencia_directores}\n")
```

#### II) ¿Quién es el director con más películas?

```python
director_repetido = raw_dataframe["director"].mode()[0] 
print(f"El director más repetido es: {director_repetido}")
```

#### III) ¿Cuántas series hay? ¿Y cuántos programas de TV?

```python
cant_programas = raw_dataframe[raw_dataframe["type"] == "TV Show"].shape[0]
print(f"La cantidad de programas es: {cant_programas}")

cant_series = raw_dataframe[raw_dataframe["listed_in"].str.contains("series", case=False)].shape[0]
print(f"La cantidad de series es: {cant_series}")
```

#### IV) ¿Cuál es la película más vieja del catálogo? ¿Y la serie?

```python
fecha_mas_vieja = raw_dataframe["release_year"].min()
peliculas_mas_viejas = raw_dataframe[raw_dataframe["release_year"] == fecha_mas_vieja]
peliculas_mas_viejas
```

#### V) Dado un país, determinar cuántas películas de ese país hay en la base de datos

```python
pais = input("Ingrese un país:")
cant_pais = raw_dataframe[raw_dataframe["country"].str.casefold() == pais.casefold()].shape[0]
print(f"El país que elegiste aparece {cant_pais} veces")
```
