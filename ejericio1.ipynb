- **GENDER:** dato cualitativo nominal. Indica el sexo de las personas y puede ayudar a encontrar una relación biológica con la obesidad. La muestra se divide casi en mitad hombres, mitad mujeres, por lo que, no hubo una preferencia de sexo.
- **CALC:** dato cualitativo nominal. Muestra si la persona consume alcohol, un hábito que puede perjudicar la salud del individuo. La respuesta más usada fue "a veces".
- **FAVC:** dato cualitativo nominal. Podría significar si consume alimentos con altas calorías con frecuencia que se puede relacionar con la ganancia de peso.
- **FCVC:** dato cuantitativo discreto. Podría referirse a la frecuencia de consumo de verduras por semana. Brinda información sobre los hábitos de la persona que podrían relacionarse con la salud. El 75% de las personas come hasta 3 veces y el 50% hasta 2, lo que podría ser un indicativo de una dieta insuficiente, siendo que el 75% de las personas come 3 veces al día (lo que nos da un total de 21 comidas a la semana).

- **SMOKE:** dato cualitativo nominal. Brinda información sobre los hábitos de la persona que podrían relacionarse con la salud. La mayoría de la gente respondió que no, dato que pareciera indicar una desconexión entre la obesidad y fumar (es sólo una mera idea).
