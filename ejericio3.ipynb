## Analizar los valores de las columnas. ¿Qué tipo de datos hay en cada una? ¿Qué representan? Incluir información del análisis en la notebook.

- **TITLE:** dato cualitativo nominal. Guarda el titulo de la pelicula o serie, puede servir para saber qué títulos son llamativos o conocidos.
- **PLATFORM:** dato cualitativo nominal. Se refiere a la plataforma de streaming donde se encuentra la serie o pelicula, lo que puede ser útil para ver el rendimiento de las plataformas. La plataforma que más aparece es Netflix
- **TYPE:** dato cualitativo nominal. Representa el tipo de producto: si es una pelicula o serie. El producto más optado en los datos son las peliculas.
- **IMDBVOTES:** dato cuantitativo discreto. Se refiere a la gente que votó en IMDB.
- **IMDBSCORE:** dato cuantitativo continuo. Indica el puntaje del filme en IMDb.
- **TMDBPOPULARITY:** dato cuantitativo continuo. Muestra un índice de popularidad.
- **TMDBSCORE:** dato cuantitativo continuo. Indica el puntaje de la película en TMDb.
- **RUNTIME:** dato cuantitativo discreto. Se refiere a la duración del producto. El promedio de películas es 1 hora y 20
- **GENRES:** dato cualitativo nominal. Indica el género de la obra, en este caso, el más recurrente es el Documental.
- **DIRECTOR:** dato cuantitativo nominal. El director podría indicarnos tendencias de la gente. El director que más aparece es Steven Spielberg
- **YEAR_RANGE:** dato cualitativo ordinal. Puede referirse al rango de años en los cuales está ambientado el producto.
- **URL:** dato cualitativo nominal.

## Proponer una serie de análisis propios:

#### I) ¿Cuáles son las películas mejor rankeadas?

Para saber cuáles son las películas mejor rankeadas deberíamos tener en cuenta la cantidad de gente que vota. No es lo mismo que voten nueve personas un diez, a que voten un millón. De esta manera:

```python
df_ranking = raw_dataframe.copy()
df_ranking['imdbScore'] = df_ranking['imdbScore'] / df_ranking['imdbVotes']
df_ranking = df_ranking.sort_values(by='imdbScore')
df_ranking_peliculas = df_ranking[df_ranking['type'] == 'MOVIE']
df_ranking_peliculas.head()
```

Como se puede observar, el ranking tiene mucho más sentido que decir que "Max Steel" es la mejor película que se encuentra en los datos (votaron seis personas):

```python
raw_dataframe = raw_dataframe.sort_values(by='imdbScore', ascending=False)
raw_dataframe.head()
```

#### II) ¿Cuáles son las series mejor rankeadas?

```python
df_ranking_series = df_ranking[df_ranking['type'] == 'SHOW']
df_ranking_series.head()
```

#### IV) ¿Cuál es es el director mejor rankeado en IMDb?

Para saber cuál es el director con mejor ranking en sus películas, podría utilizar las funciones de grupo:

```python
df_ranking_promedio = df_ranking.groupby('director')['imdbScore'].mean()
df_ranking_promedio = df_ranking_promedio.sort_values()

print(f"Como verás el director mejor rankeado es (el del imperio contraataca): {df_ranking_promedio.idxmin()}")
df_ranking_promedio.head()
```
